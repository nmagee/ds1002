# Practice with Viewing, Filtering & Grouping

Useful tools for seeing broadly across your data set

```python
# set up pandas
import pandas as pd
```

```python
# import your dataset
rides = pd.read_csv('https://ds1002-resources.s3.amazonaws.com/data/ride_sharing.csv')
rides.head(3)
```

```python
# view only the records based upon a given value of a given column.
# look for all rows where station_A_id is equal to 81.

# To do this, use the .query() method against your dataframe. As a parameter
# set the "station_A_id == 81" .
rides.query("station_A_id == 81")
```

```python
# We can also filter for string values with .query()
rides.query('user_gender == "Female"')
```
