```python
import seaborn as sns
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
df.head(5)
```

```python
filtered_df = df.loc[df['Age'] > 10]
# Select specific columns by name
selected_columns = filtered_df[['Age', 'Fare']]
# Select specific columns using loc or iloc
selected_columns = filtered_df.loc[:,['Age', 'Fare']]
print(selected_columns.head())
```

```python
sns.scatterplot(x='Age', y='Fare', data=filtered_df, hue='Fare', size='Fare', marker='x')
```
